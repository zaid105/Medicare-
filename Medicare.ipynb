## Hypothesis

#### Test null Average avg_Mdcr_Pymt_Amt for Place_of_srvc is equal 

```python
df.Avg_Mdcr_Pymt_Amt.groupby(df.Place_Of_Srvc).mean()
```

```python
df.Avg_Mdcr_Pymt_Amt.groupby(df.Place_Of_Srvc).var()
```

```python
# Splitting
Fclass = df[df.Place_Of_Srvc == "F"] 
Oclass = df[df.Place_Of_Srvc == "O"] 
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(Fclass.Avg_Mdcr_Pymt_Amt,Oclass.Avg_Mdcr_Pymt_Amt)
# Since pvalue=1.0331351382064729e-25 is less than 0.05 - Reject null
```

####  Test null Average Tot_bane_day_srvc for place-of_srvc is equal

```python
df.Tot_Bene_Day_Srvcs.groupby(df.Place_Of_Srvc).mean()
```

```python
# Splitting
Fclass = df[df.Place_Of_Srvc == "F"] 
Oclass = df[df.Place_Of_Srvc == "O"] 
```

```python
ttest_ind(Fclass.Tot_Bene_Day_Srvcs,Oclass.Tot_Bene_Day_Srvcs)
# Since pvalue=1.0331351382064729e-25 is less than 0.05 - Reject null
```

####  Test null no assosiation between Rndrng_prvdng_Geo_lvl and place_of_srvc

```python
from scipy.stats import chi2_contingency
```

```python
chi2_contingency(pd.crosstab(df.Rndrng_Prvdr_Geo_Lvl,df.Place_Of_Srvc))
# Since pvalue=0.548014490212158 is greater than 0.05 - Fail to reject null
```

### Model Building

#### Linear Rigression

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression().fit(X,y)
```

```python
LR.score(X,y)
```

```python
regpred = LR.predict(X)
```

```python
np.sqrt(np.mean((y-regpred)**2)) # RMSE
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
DT = DecisionTreeRegressor(max_depth=8).fit(X,y)
DT.score(X,y)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
v = cross_val_score(DT,X,y)
```

```python
v.mean()
```

#### Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF = RandomForestRegressor().fit(X,y)
```

```python
RF.score(X,y)
```

#### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
GB = GradientBoostingRegressor(n_estimators=500).fit(X,y)
```

```python
GB.score(X,y)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
gbpred = GB.predict(X)
```

```python
print("MAE = ",mean_absolute_error(y,gbpred))
```

```python
print("MSE = ", mean_squared_error(y,gbpred))
```

```python
print("RMSE = ",np.sqrt((mean_squared_error(y,gbpred))))
```

```python
r2 = (r2_score(y,gbpred))
print("R2 score",r2)
```
